# EXTRACCION

```python
# Cargar las librerías a utilizar
import pandas as pd
import json
```

```python
# Colocar la ruta a los datos en una variable
ruta_archivo = "C:/Users/Usuario/Desktop/Varios yo/SoyHenry/P1/Henry_PI1_Steam/data/output_steam_games.json"

# Lista para almacenar los diccionarios individuales
lista_datos = []

# Abrir el archivo y cargar cada línea como un objeto JSON
with open(ruta_archivo, 'r') as archivo:
    for linea in archivo:
        try:
            # Cargar el JSON de cada línea y agregarlo a la lista
            diccionario_datos = json.loads(linea)
            lista_datos.append(diccionario_datos)
        except json.JSONDecodeError as e:
            print(f"Error de decodificación JSON en línea: {e}")

# Crear un DataFrame a partir de la lista de diccionarios
df_games = pd.DataFrame(lista_datos)

# Mostrar el DataFrame
df_games.tail(5)
```

```python
print(df_games.shape)
df_games.info()
```

Realizo una primera transformación realmente necesaria en este momento, eliminando las filas que están completamente vacías antes de generar el archivo .CSV

```python
# Eliminar las filas vacías
df_games.dropna(how="all", inplace=True)
```

```python
# Exportar el Dataframe a .CSV
df_games.to_csv("data/steam_games.csv", index=False)
```

# TRANSFORMACION

```python
# Cargar las librerías a utilizar
import pandas as pd
```

```python
# Cargar el .csv generado en el paso anterior
df_games = pd.read_csv("data/steam_games.csv")
```

```python
# Revisar la presencia de nulos y nombres de columnas
print(df_games.isna().sum())
```

```python
# Revisar la cantidad de filas del Dataframe
df_games.shape
```

```python
# Aplicar drop_duplicates para ver si hay valores duplicados
df_games.drop_duplicates()
```

```python
# Revisar nuevamente la cantidad de filas. No había duplicados. 
df_games.shape
```

```python
# Eliminar las columnas innecesarias
df_games = df_games.drop('url',axis=1) # No voy a utilizar el URL del juego en Steam para ningún análisis
df_games = df_games.drop('reviews_url',axis=1) # Las reviews están en otro Dataset
df_games = df_games.drop('publisher',axis=1) # No voy a utilizar el publicista para ningún análisis
df_games = df_games.drop('title',axis=1) # Es redundante con app_name pero tiene más nulos
```

```python
# Convertir el tipo de la columna 'release_date' a fecha con el formato 'YYYY-MM-DD'
df_games['release_date'] = pd.to_datetime(df_games['release_date'], format='%Y-%m-%d', errors='coerce')
```

### Aún tenemos varias columnas con datos anidados, que se procede a desanidar

GENRES

```python
# Convertir 'genres' en string
df_games['genres'] = df_games['genres'].astype(str)

# Desanidar la columna 'genres' y crear una nueva columna 'genre'
df_games['genre'] = df_games['genres'].apply(lambda x: x.strip("[]").replace("'", "").split(', '))

# Explotar la columna 'genre' en múltiples filas
df_games = df_games.explode('genre')

# Eliminar la columna 'genres' original
df_games = df_games.drop('genres',axis=1)
```

TAGS

```python
# Convertir 'tags' en string
df_games['tags'] = df_games['tags'].astype(str)

# Desanidar la columna 'tags' y crear una nueva columna 'tag'
df_games['tag'] = df_games['tags'].apply(lambda x: x.strip("[]").replace("'", "").split(', '))

# Explotar la columna 'tag' en múltiples filas
df_games = df_games.explode('tag')

#Eliminar la 'tags' original
df_games = df_games.drop('tags',axis=1)
```

SPECS

```python
#Convertir la columna 'specs' en string
df_games['specs'] = df_games['specs'].astype(str)

# Desanidar la columna 'specs' y crear una nueva columna 'spec'
df_games['spec'] = df_games['specs'].apply(lambda x: x.strip("[]").replace("'", "").split(', '))

# Explotar la columna 'spec' en múltiples filas
df_games = df_games.explode('spec')

# Eliminar la 'specs' original
df_games = df_games.drop('specs',axis=1)
```

```python
# Exportar a un nuevo .csv el Dataframe transformado
df_games.to_csv("C:/Users/Usuario/Desktop/Varios yo/SoyHenry/P1/Henry_PI1_Steam/data/steam_games_2.csv")
```
