# EXTRACCION 

```python
# Cargar las librerías a utilizar
import pandas as pd
import json
```

```python
# Ruta al archivo .JSON
ruta_archivo = "data/australian_user_reviews.json"

# Inicializar una lista para almacenar los DataFrames
rows = []

# Leer el archivo JSON línea por línea y normalizar cada línea
with open(ruta_archivo, "r", encoding="utf-8") as archivo:
    for linea in archivo:
        # Utilizar json_normalize para convertir cada línea en un DataFrame
        row = pd.json_normalize(eval(linea))
        # Agregar el DataFrame a la lista
        rows.append(row)

# Concatenar todos los DataFrames en un DataFrame final
df_reviews = pd.concat(rows, ignore_index=True)

# Revisar si el Dataframe está correcto
df_reviews.head()
```

```python
# Revisar el formato de la última columna. Cada elemento tiene varias listas con datos del review
df_reviews.loc[0,"reviews"]
```

```python
# Separar los distintos reviews
df_reviews = df_reviews.explode("reviews")

df_reviews.head()
```

```python
# Convertir la columna "reviews" en un DataFrame independiente
df_temp = df_reviews["reviews"].apply(pd.Series)

df_temp.head()
```

```python
# Unir el DataFrame original con el nuevo DataFrame de reviews
df_reviews = pd.concat([df_reviews, df_temp], axis=1)

df_reviews.head()
```

```python
# Eliminar la columna original "reviews"
df_reviews.drop("reviews", axis=1, inplace=True)

df_reviews.head()
```

```python
# Exportar el Dataframe a .CSV
df_reviews.to_csv("data/user_reviews.csv", index=False)
```

# TRANSFORMACION

```python
# Cargar las librerías a utilizar
import pandas as pd
from textblob import TextBlob
```

```python
# Cargar el .csv generado en el paso anterior
df_reviews = pd.read_csv("data/user_reviews.csv")
```

```python
# Revisar la cantidad de filas del Dataframe
df_reviews.shape
```

Se necesita llevar la columna de fecha de posteo a un formato acorde a las necesidades del análisis. Para el mismo, solo puede interesar el año de posteo, por lo que se va a trabajar la columna para obtener el año. En el caso de fechas de posteo en las cuales no hay año, no se guardará ningún valor.

```python
# Aplicar drop_duplicates para ver si hay filas duplicadas
df_reviews.drop_duplicates()
```

```python
# Revisar nuevamente la cantidad de filas. No había filas duplicadas
df_reviews.shape
```

```python
# Dividir la columna 'posted' por la coma y seleccionar la segunda parte
df_reviews['posted'] = df_reviews['posted'].str.split(',').str[1]
```

```python
# Eliminar el punto del final
df_reviews['posted'] = df_reviews['posted'].str.replace('.', '', case=False)
```

```python
# Eliminar espacios en blanco en caso de que los haya
df_reviews['posted'] = df_reviews['posted'].str.strip()
```

```python
df_reviews = df_reviews.drop('user_url',axis=1) # No interesa el URL del usuario para el análisis
df_reviews = df_reviews.drop('last_edited',axis=1) # Solo usaremos el año del posteo original
df_reviews = df_reviews.drop('funny',axis=1) # El análisis de sentimiento se realiza con el texto del posteo
df_reviews = df_reviews.drop('helpful',axis=1) # No interesa el posteo en relación a la ayuda para otros usuarios
df_reviews = df_reviews.drop('0',axis=1) # Columna vacía
```

En las siguientes líneas de código se va a realizar el análisis de sentimiento con la librería textblob, y se van a crear las dos columnas recomendadas, una que muestre la polaridad resultante y la otra con la calificación del sentimiento, siendo 0 para reviews negativas, 1 para reviews neutras y 2 para reviews positivas.

```python
# Convertir todos los valores de la columna a string
df_reviews['review'] = df_reviews['review'].astype(str)
```

```python
# Crear la columna 'polarity' y pasarle la función que calcula la polaridad del sentimeinto
df_reviews['polarity'] = df_reviews['review'].apply(lambda text: TextBlob(text).sentiment.polarity)
```

```python
# Reviso las estadísticas de polaridad para definir los intérvalos de categoría de review
df_reviews["polarity"].describe()
```

La definición de los intérvalos es una cuestión de criterio personal. La mayoría de los reviews tienen una polaridad que está entre -0.20 y 0.40. En este caso elijo como review negativa cualquier valor por debajo de cero, como review neutra los valores = 0.00 (las reviews inexistentes (valor cero) se considerarán neutras), y cualquier valor por encima de cero lo considero como review positiva.

(Obs: Estos valores para los parámetros fueron adoptados después de varias iteraciones)

```python
# Crear la columna 'sentiment' considerando los intérvalos adoptados
df_reviews['sentiment'] = pd.cut(df_reviews['polarity'], bins=[-float('inf'), -0.001, 0.0, float('inf')], labels=[0, 1, 2])
```

```python
# Reviso la distribución del sentimiento para comprobar la segmentación adoptada
df_reviews["sentiment"].value_counts()
```

```python
# Eliminar la columna 'review', de la cual ya se ha obtenido lo necesario
df_reviews = df_reviews.drop('review', axis = 1)
```
