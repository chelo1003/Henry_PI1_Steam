Para esta función se necesita:

- Desde el Dataframe steam_games el id y el nombre de los juegos.
- Desde el Dataframe users_reviews el id de juego, la fecha de posteo, la columna de recomendación y la columna de sentimiento.

Con estos datos se crea el Dataframe que se sube para la consulta

```python
# Importar las librerías a utilizar
import pandas as pd
```

```python
# Cargar los Dataframes
df_user_reviews = pd.read_csv("C:/Users/Usuario/Desktop/Varios yo/SoyHenry/P1/Henry_PI1_Steam/data/user_reviews_2.csv")
df_steam_games = pd.read_csv("C:/Users/Usuario/Desktop/Varios yo/SoyHenry/P1/Henry_PI1_Steam/data/steam_games_2.csv")
```

## TRANSFORMACIONES de steam_games

```python
# Revisar las columnas de steam_games
df_steam_games.columns
```

```python
# Desde el Dataframe df_steam_games eliminar las columnas que no se necesitan
df_steam_games.drop("Unnamed: 0", axis=1, inplace=True)
df_steam_games.drop("release_date", axis=1, inplace=True)
df_steam_games.drop("price", axis=1, inplace=True)
df_steam_games.drop("early_access", axis=1, inplace=True)
df_steam_games.drop("developer", axis=1, inplace=True)
df_steam_games.drop("genre", axis=1, inplace=True)
df_steam_games.drop("tag", axis=1, inplace=True)
df_steam_games.drop("spec", axis=1, inplace=True)
df_steam_games
```

```python
# Debido a la presenca de "tag" y "spec" ahora eliminadas, quedaron muchos duplicados
df_steam_games.drop_duplicates(inplace=True)
```

```python
# Revisar el Dataframe
df_steam_games.info()
```

## TRANSFORMACIONES de user_reviews

```python
# Revisar las columnas de user_reviews
df_user_reviews.columns
```

```python
# Desde el Dataframe df_user_reviews eliminar las columnas que no se necesitan
df_user_reviews.drop("user_id", axis=1, inplace=True)
df_user_reviews.drop("polarity", axis=1, inplace=True)
df_user_reviews
```

```python
df_user_reviews.rename(columns={'item_id': 'id'}, inplace=True)
```

--------------------------

## Ahora con los Dataframes procesados, creo el csv para la consulta

```python
# Quitar de user_reviews los registros que no recomienden o que tengan reseñas negativas
df_recomendaciones = df_user_reviews[(df_user_reviews['recommend'] == True) & ((df_user_reviews['sentiment'] == 1) | (df_user_reviews['sentiment'] == 2))]
```

```python
df_recomendaciones
```

```python
# Eliminar las columnas 'recomended' y 'sentiment' porque ya se aplicó el filtro
df_recomendaciones.drop("recommend", axis=1, inplace=True)
df_recomendaciones.drop("sentiment", axis=1, inplace=True)
df_recomendaciones
```

```python
# Revisar los valores de la fecha de posteo
df_recomendaciones['posted'].unique()
```

(Como en este caso la función va a trabajar para un año elegido por el consumidor de la API, no tiene sentido trabajar con registros sin año, y se van a eliminar)

```python
# Borrar los NaN de la fecha de posteo
df_recomendaciones.dropna(subset=['posted'], inplace=True)
```

```python
df_recomendaciones
```

```python
# Revisar los valores de la fecha de posteo
df_recomendaciones['posted'].unique()
```

```python
# Convertir a tipo de dato entero la columna del año de posteo
df_recomendaciones['posted'] = df_recomendaciones['posted'].astype(int)
df_recomendaciones
```

```python
# Contar para cada año la cantidad de veces que un juego fue recomendado o tiene un posteo con sentimiento favorable
df_recomendaciones = df_recomendaciones.groupby(['posted', 'id']).size().reset_index(name='conteo')
```

```python
df_recomendaciones
```

```python
# Ordenar el Dataframe por año y conteo
df_recomendaciones = df_recomendaciones.sort_values(['posted', 'conteo'], ascending=[True, False])
```

```python
df_recomendaciones
```

```python
# Extraer solo los top 3 juegos por conteo para cada año
df_recomendaciones = df_recomendaciones.groupby('posted').apply(lambda x: x.nlargest(3, 'conteo')).reset_index(drop=True)
```

```python
df_recomendaciones
```

```python
# Unir los Dataframes de recomendaciones y steam_games para poner el nombre de los juegos
df_recomendaciones = df_recomendaciones.merge(df_steam_games, on=['id'], how="left")
```

```python
df_recomendaciones
```

```python
# Eliminar la columna con el id del juego porque ya se agregó el nombre
df_recomendaciones.drop("id", axis=1, inplace=True)
df_recomendaciones
```

```python
# Exportar el Dataframe a .csv eliminando el índice
df_recomendaciones.to_csv("UsersRecommend.csv", index=False)
```
